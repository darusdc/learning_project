```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from scipy.stats import norm
```

```python
x=np.random.normal(6,3,size=1000)
x.std()
```

```python
series=pd.Series(x)
series.plot.density()
```

```python
series.describe()
```

```python
plt.axvline(series.mean(),color='red')
plt.axvline(series.median(),color='blue')
series.plot.density()
```

## SkewTest

```python
stats.skewtest(series)
```

## Kurtosis Test

```python
stats.kurtosistest(series)
```

```python
plt.axvline(series.mean(),color='red')
plt.axvline(series.mode().sort_values()[499])
plt.axvline(series.median(),color='blue')
series.plot.density()
```
