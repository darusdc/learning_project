```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cleo = pd.read_csv('CLEO_JK.csv')
icpb=pd.read_csv('ICBP_JK.csv')
myor=pd.read_csv('MYOR_JK.csv')
ultj=pd.read_csv('ULTJ_JK.csv')
```

```python
cleo.head(5)
```

```python
cleo['Date']=pd.to_datetime(cleo['Date'])
ultj['Date']=pd.to_datetime(ultj['Date'])
myor['Date']=pd.to_datetime(myor['Date'])
icpb['Date']=pd.to_datetime(icpb['Date'])
```

```python
plt.figure(figsize=(10,6))
x=cleo['Date']
y=cleo['Close']-cleo['Open']
plt.plot(x,y,'Black')
plt.title('Saham Cleo',size=12)
plt.xlabel('month')
plt.ylabel('Price')
```

```python
plt.figure(figsize=(10,6))
x=cleo['Date']
y1=cleo['Close']
y2=ultj['Close']
y3=icpb['Close']
y4=myor['Close']
plt.plot(x,y1,'Blue')
plt.plot(x,y2, 'Red')
plt.plot(x,y3, 'Yellow')
plt.plot(x,y4,'Green')
plt.xlabel('harian')
plt.ylabel('harga')
plt.legend(['cleo','ultj','icbp','myor'])
plt.show()
```

```python
plt.figure(figsize=(10,6))
x=ultj['Date']
y=ultj['Close']-ultj['Open']
plt.plot(x,y,'Black')
plt.title('Saham Ultj',size=12)
plt.xlabel('month')
plt.ylabel('Price')
```

```python
x=cleo['Date']
y1=cleo['Open']
y2=ultj['Open']
y3=myor['Open']
y4=icpb['Open']

fig=plt.figure(figsize=(15,7))
fig.suptitle('Histori harga saham consumen goods 2019-2020')

ax1=fig.add_subplot(2,2,1)
ax1.set_title('Cleo')
ax1.plot(x,y1,color='blue')

ax2=fig.add_subplot(2,2,2)
ax2.set_title('ULTJ')
ax2.plot(x,y2,color='red')

ax3=fig.add_subplot(2,2,3)
ax3.set_title('MYOR')
ax3.plot(x,y3,color='yellow')

ax4=fig.add_subplot(2,2,4)
ax4.set_title('ICBP')
ax4.plot(x,y4,color='green')

plt.show()
```

# Stem Plots

```python
x=cleo['Date']
cleo_stem=cleo['Open'].diff()
icbp_stem=icpb['Open'].diff()
myor_stem=myor['Open'].diff()
ultj_sten=ultj['Open'].diff()
```

```python
plt.figure(figsize=(15,16))
#cleo
plt.title("CLEO")
plt.subplot(4,1,1)
plt.stem(x,cleo_stem,linefmt='r--',markerfmt='b-')
#icbp
plt.title('ICBP')
plt.subplot(4,1,2)
plt.stem(x,icbp_stem,linefmt='r--',markerfmt='b-')
#MYOR
plt.title('MYOR')
plt.subplot(4,1,3)
plt.stem(x,myor_stem,linefmt='r--',markerfmt='b-')
#ultj
plt.title('ULTJ')
plt.subplot(4,1,4)
plt.stem(x,ultj_sten,linefmt='r--',markerfmt='b-')

plt.show()
```

# Scatter Plot

```python
nba=pd.read_csv('nba.csv')
nba
```

```python
nba['Height']=nba['Height'].str.replace('-','.')
nba['Height']=nba['Height'].astype(float)
```

```python
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)
plt.scatter(nba['Age'],nba['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Correlation age and salary')

plt.subplot(2,2,2)
plt.scatter(nba['Number'],nba['Salary'])
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Correlation Number and salary')

plt.show()
```
