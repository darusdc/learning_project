```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import requests
```

### Import Data

```python
#bank
bni=pd.read_csv('BBNI.JK.csv')
bri=pd.read_csv('BBRI.JK.csv')
mandiri=pd.read_csv('BMRI.JK.csv')
btpn=pd.read_csv('BTPN.JK.csv')
bca=pd.read_csv('BBCA.csv')
bank_grup=[bni,bri,mandiri,btpn,bca]
#Property
apl=pd.read_csv('APLN.JK.csv')
alsut=pd.read_csv('ASRI.JK.csv')
bsd=pd.read_csv('BSDE.JK.csv')
citra=pd.read_csv('CTRA.JK.csv')
property_grup=[apl,alsut,bsd,citra]
```

### Create Lineplot

Bank BNI

```python
x=bni['Date']
y=bni['Close']
plt.figure(figsize=(10,6))
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.title("Bank BNI Performance from Agustus 3, 2019 - Agustus 3,2020")
plt.tick_params(axis='x',which='both', bottom=False,top=False,labelbottom=False)
plt.plot(x,y,color='blue')
```

SBS Price all Bank

```python
x=bni['Date']
y1=bni['Close']
y2=bri['Close']
y5=mandiri['Close']
y3=btpn['Close']
y4=bca['Close']
fig=plt.figure(figsize=(10,10))
fig.suptitle("Bank Performance from 2019-2020 (august)")
#BNI
ax1=fig.add_subplot(3,2,1)
ax1.set_title("BNI")
ax1.plot(x,y1,color='blue')
#BRI
ax2=fig.add_subplot(3,2,2)
ax2.set_title("BRI")
ax2.plot(x,y2,color='red')
#Mandiri
ax3=fig.add_subplot(3,2,3)
ax3.set_title("Mandiri")
ax3.plot(x,y5,color='green')
#BTPN
ax4=fig.add_subplot(3,2,4)
ax4.set_title("BTPN")
ax4.plot(x,y3,color='yellow')
#BCA
ax5=fig.add_subplot(3,2,5)
ax5.set_title("BCA")
ax5.plot(bca['Date'],y4,color='green')

plt.show()
```

```python
x=bni['Date']
y1=bni['Close']
y2=bri['Close']
y5=mandiri['Close']
y3=btpn['Close']
y4=bca['Close']
plt.plot(x,y1,label='BNI')
plt.plot(x,y2,label='BRI')
plt.plot(x,y3,label='BTPN')
plt.plot(bca['Date'],y4,label='BCA')
plt.plot(x,y5,label='Mandiri')

plt.legend()
plt.grid()
plt.figure(figsize=(16,9))
```

## Hubungan Korelasi antara 

```python
bni_stem=bni['Adj Close']
apl_stem=apl['Adj Close']
plt.scatter(bni_stem,apl_stem)
plt.title('BNI x APL')
plt.xlabel('APL')
plt.ylabel('BNI')
```

```python
plt.figure(figsize=(20,10))
plt.suptitle('Korelasi harga saham bank dengan property')
#APL x BNI
plt.subplot(2,3,1)
plt.title('APL x BNI')
plt.scatter(bni_stem,apl_stem)
plt.xlabel('BNI')
plt.ylabel('APL')
# Ciputra x Mandiri
plt.subplot(2,3,2)
plt.title('Ciputra x Mandiri')
plt.scatter(mandiri['Adj Close'],citra['Adj Close'])
plt.xlabel('Mandiri')
plt.ylabel('Ciputra')
# BSD x BTPN
plt.subplot(2,3,3)
plt.title('BSD x BTPN')
plt.scatter(btpn['Adj Close'],bsd['Adj Close'])
plt.xlabel('BTPN')
plt.ylabel('BSD')
```
